## 콘솔 출력의 기본
-----------------------------------
제일 먼저 배울 내용은 콘솔 출력입니다.
>'콘솔'은 컴퓨터와 소통하는 창입니다.

프로그래밍 초보자라면 컴퓨터의 생각 구조가 익숙하지 않기 때문에 콘솔에 출력되는 결괏값을 자주 확인해야합니다. 만약 여러분이 계산기를 제작한다고 가정해봅시다. 계산기의 기능을 모두 구현해도 결과 화면이 없다면 연산 결괏값을 보여주지 못하여 이 꼐산기가 제대로 동작하는지 알 수 없습니다. 덧셈 기능 하나만 구현했더라도 기능이 제대로 작동함을 알기 위해 '콘솔 출력'을 해야 합니다.

계산기처럼 필요한 기능을 구현한 뒤 입력값을 넣을 때 콘솔을 통해 확인할 수 있고, 기능에 따라 계산된 결과값 또한 콘솔 창으로 확인합니다.
이렇게 콘솔에 값을 입력, 출력할 수 있으며, 이를 위해 콘솔 입출력 함수를 사용합니다. 그 중 콘솔 출력 함수에 대해 먼저 알아봅시다.

```python
print(10)
print(10, 20)
```

파이썬에서 콘솔 출력을 하기 위해 제일 많이 쓰이는 print() 부터 사용해봅시다.

사용 형태는
> print( 출력하고자 하는 데이터 )

이며, 여러 데이터를 나열할 때는, 콤마(,)를 사용합니다.

## 변수의 선언과 초기화
-----------------------------------
변수에 대해서 알아봅시다!
변수는 데이터(정보)를 다루는 데 있어서 가장 기초로 알아야 하지만, 그만큼 잘 사용하기 매우 힘든 요소입니다.

> 변수는 말 그대로 변하는 수이자, "특정한 형식을 가진 값을 저장하는 공간"을 의미합니다.

일반적으로 변수는 세 가지 요소와 함께 생성됩니다.
* 자료형(data type): 변수가 가지는 특정한 형식(1은 정수형, c는 문자형 등등)
* 변수명: 변수의 이름
* 변숫값: 변수에 저장한 값

세 요소 중 자료형과 변수명을 작성하여 변수를 생성하면
> '변수를 선언했다'

라고 말하며, 새로 생성된 변수에 처음 값을 입력하는 과정을
> '변수 초기화'

라고 합니다. 그렇다면 다음 세 문장은 모두 동일한 뜻을 가집니다.

+ int x = 2
+ int형(정수형)이며 x라는 변수명을 가진 변수에 2라는 변숫값을 할당한다
+ int형(정수형)으로 선언된 변수 x를 변숫값 2로 초기화 한다.

C언와 같은 엄격한 컴퓨터 언어에서는 변수에 저장할 값에 따라 '자료형'을 직접 작성해야 합니다.
때문에
> 'int b = 10.5'

와 같은 잘못된 문장을 작성할 가능성이 있습니다.
하지만 파이썬에서는 비교적 이런 실수를 할 가능성이 적습니다.

```python
# 자료형 입력이 필요 없는 파이썬
a = 10
b = 10.5
print(a, b)
```

a와 b 앞에 그 어떤 자료형도 보이지 않지만 코드가 정상 실행됩니다.

왜 그럴까요?

> 파이썬은 자료형(데이터의 타입, 형식)을 자동으로 추론합니다.

즉, 변수 이름과 변수 값만 있으면 변수가 정상 생성되기 때문에 위 코드는 숫자형 변수 a에 변숫값 10이, 실수형 변수 b에 변숫값 10.5가 초기화되었다고 해석합니다.

그렇다면 만약 선언하고 싶은 변수가 모두 정수형일 때는 어떻게 선언해야 할까요?
동일한 자료형을 가진 변수들은 한 문장의 코드로 한 번에 초기화 할 수 있습니다.

```python
a, b = 10, 30
print(a, b)
```

```python
a = 10
b = 30
print(a, b)
```

위의 두 코드의 결과가 a는 10, b는 30으로 값이 동일하게 초기화 된 것을 확인하셨나요?
실제로 많은 사람이 사용 중인 다양한 프로그램들은 엄청난 길이의 코드로 작성되어 있습니다.

그 때문에 코드를 한 줄 줄이는 것만으로도 가독성이 달라지므로 이러한 간결화 작업은 미리 습관화하는 것을 추천해 드립니다.

```python
a = "Hello"
b = "Kim!"
c = "10"
d = 10
e = b

result = a + b
d = 5
print(a, b, c, d, e, result)
```

마지막으로 숫자 이외에 다른 자료형으로 초기화되는 변수를 살펴봅시다.

예시 코드의 변수는 a, b, c, d, e 순서대로
문자열, 문자열, 문자열, 숫자, 문자열의 자료형을 갖습니다.

위 코드에서 눈여겨볼 점에 대해 하나씩 짚어봅시다.

1. 3번째 줄에서 c는 왜 문자열일까요?
    => 큰따옴표("")는 문자열을 표현하기 위한 장치입니다.
2. 변수는 변수로 초기화가 가능합니다.
3. result = a + b 와 같이 연산을 포함하여 초기화 할 수 있습니다.
4. 문자열을 + 연산할 수 있습니다.
5. 변수 안에 저장되는 변숫값은 동일한 자료형의 다른 값으로 변경하여 사용할 수 있습니다.
